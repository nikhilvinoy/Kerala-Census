```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## Reading Census Data 

```python
census_data  = pd.read_csv(r"C:\Users\GODFATHER\Desktop\Data Analytics\CensusData\india-districts-census-2011.csv")
census_data
```

```python
kerala_census  = census_data[census_data["State Name"] =="KERALA"]
```

```python
kerala_census
```

## Districtwise Population 

```python
plt.subplots(figsize = (15,8))
ct = kerala_census.groupby('District name')['Population'].sum()
ax = ct.sort_values(ascending=False).plot(kind="bar")
for p in ax.patches:
    ax.annotate(format(p.get_height()), (p.get_x()+0.1, p.get_height()+2),fontsize=10)
plt.show()
```

Malappuram is the most populated district in Kerala followed by Thiruvananthapuram & Ernakulam.

## Religions in Kerala 

```python
IRPdf  = kerala_census[['Hindus','Muslims','Christians','Sikhs','Buddhists','Jains','Others_Religions']].sum().head(3)
IRPdfg = IRPdf.plot(kind="pie",radius=2,autopct='%1.1f%%', startangle=90)
IRPdfg.set_title("Religions in Kerala",loc="center",fontdict={'fontsize':20},position=(0.5,1.55))
```

Hinduism is the most followed religion in Kerala followed by Islam.

## AgeGroup Distribution Districtwise 

```python
cen0_29 = kerala_census.groupby("District name").mean()
Age0_29=kerala_census[['Age_Group_0_29']].to_dict()
Y09=["0-29" for i in range(14)]
S0_29=list(kerala_census["District name"])
P0_29=list(Age0_29['Age_Group_0_29'].values())


cen30_49 = kerala_census.groupby("District name").mean()
Age30_49=kerala_census[['Age_Group_30_49']].to_dict()
Y30=["30-49" for i in range(14)]
S30_49=list(kerala_census["District name"])
P30_49=list(Age30_49['Age_Group_30_49'].values())


cen50 = kerala_census.groupby("District name").mean()
Age50=kerala_census[['Age_Group_50']].to_dict()
Y50=["50+" for i in range(14)]
S50=list(kerala_census["District name"])
P50=list(Age50['Age_Group_50'].values())


Data = {'Age_Group':Y09+Y30+Y50,'Districts':S0_29+S30_49+S50,'Population':P0_29+P30_49+P50}
DF = pd.DataFrame(data=Data)
ax = plt.subplots(figsize=(20, 20))
sns.barplot(y=DF.Population,x=DF.Districts,hue=DF.Age_Group)
plt.show()
```

Young population (0-29) is more in Malappuram , whereas the above 50+ population is high in Ernakulam.
